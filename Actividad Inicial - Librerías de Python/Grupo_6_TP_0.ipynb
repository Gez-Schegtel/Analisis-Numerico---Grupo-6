## NumPy

**1) Propósito principal**  
El propósito principal de NumPy (Numerical Python) es proporcionar a Python la capacidad de realizar cálculos numéricos y científicos de manera eficiente, especialmente con grandes volúmenes de datos. Para lograrlo, introduce un objeto fundamental: el array multidimensional o `ndarray`.

Este objeto `ndarray` es la piedra angular de la librería y ofrece ventajas significativas sobre las listas nativas de Python para operaciones matemáticas:

* **Rendimiento**: Los arrays de NumPy son mucho más rápidos. Las operaciones se ejecutan en código C compilado, lo que elimina los cuellos de botella de rendimiento del intérprete de Python, permitiendo cálculos a gran velocidad.

* **Eficiencia de memoria**: Los arrays de NumPy almacenan los datos de una manera más compacta y eficiente que las listas de Python, lo que es crucial cuando se trabaja con conjuntos de datos masivos.

* **Funcionalidad matemática**: Proporciona un amplio conjunto de funciones matemáticas de alto nivel para realizar operaciones complejas sobre los arrays de forma sencilla y directa. Esto incluye desde operaciones aritméticas básicas hasta funciones de álgebra lineal, transformadas de Fourier y generación de números aleatorios.

* **Conveniencia**: Permite realizar operaciones sobre arrays completos sin necesidad de escribir bucles for explícitos. Esto no solo hace el código más rápido, sino también más conciso, legible y menos propenso a errores.

**2) Estructuras/objetos que introduce**  
NumPy introduce la estructura de datos llamada **`ndarray`**, que es un **array multidimensional** de alto rendimiento. "*`ndarray`*" son las siglas de *N-dimensional array*.

Este único objeto es la base para representar:

* **Vectores:** Arrays de una dimensión.
* **Matrices:** Arrays de dos dimensiones.
* **Tensores:** Arrays de tres o más dimensiones.

A diferencia de las listas de Python, todos los elementos de un `ndarray` deben ser del **mismo tipo de dato** (por ejemplo, todos enteros o todos números de punto flotante). Esta homogeneidad es clave para la eficiencia y velocidad de NumPy.

Es importante destacar que NumPy **no maneja expresiones simbólicas**. Para la computación simbólica (manipular variables y expresiones en lugar de valores numéricos), se utiliza otra librería llamada **SymPy**.

**3) Funciones/métodos útiles**  
Dos funciones de NumPy (o de su ecosistema, como SciPy) que son extremadamente útiles en Análisis Numérico son `numpy.linalg.solve` para resolver sistemas de ecuaciones lineales y `numpy.polyfit` para la aproximación de datos mediante polinomios.

1. `numpy.linalg.solve`: Esta función resuelve sistemas de ecuaciones lineales de la forma $Ax = b$, donde **A** es una matriz de coeficientes (cuadrada), **x** es el vector de incógnitas y **b** es el vector de términos independientes. Esta es una de las tareas más fundamentales en la ciencia y la ingeniería.
  
  **¿Por qué es útil?**
  En lugar de calcular manualmente la inversa de la matriz A (lo cual es computacionalmente costoso e inestable), `np.linalg.solve` utiliza métodos numéricos eficientes y estables (como la descomposición LU) para encontrar el vector de incógnitas **x**.

2. `numpy.polyfit`: Esta función se usa para el **ajuste de curvas** (o regresión polinómica). Dado un conjunto de puntos de datos `(x, y)`, `np.polyfit` encuentra los coeficientes de un polinomio del grado que especifiques que mejor se ajusta a esos datos, minimizando el error cuadrático.

  **¿Por qué es útil?**
  Permite encontrar un modelo matemático simple (un polinomio) que describe la tendencia de un conjunto de datos experimentales o ruidosos. Esto es fundamental para la interpolación, la predicción y el análisis de tendencias.

**4) Situación práctica**  
Resolver rápidamente un sistema **Ax=B**. Supongamos que tenemos el siguiente sistema de ecuaciones:

$3x_1 + x_2 = 9$

$x_1 + 2x_2 = 8$

Podemos representarlo en forma matricial como:

$$
\begin{pmatrix} 3 & 1 \\ 1 & 2 \end{pmatrix}
\begin{pmatrix} x_1 \\ x_2 \end{pmatrix} =
\begin{pmatrix} 9 \\ 8 \end{pmatrix}
$$

Donde **A** es la matriz de los coeficientes y **B** la matriz de los términos independientes:

Otra situación donde se recurre regularmente a esta librería es cuando se necesita ajustar una curva polinómica a datos experimentales para estimar valores intermedios.

> *A estas dos situaciones prácticas las resolveremos en el bloque de código que sigue.*

````python
# ~~~~~~~ Ejemplo práctico n.°1 utilizando NumPy ~~~~~~~ #

## Resolvemos el sistema de ecuaciones 3.x1 + x2 = 9 ; x1 + 2.x2 = 8

import numpy as np

# Matriz de coeficientes A
A = np.array([[3, 1],
              [1, 2]])

# Vector de términos independientes b
b = np.array([9, 8])

# Resolvemos para x
x = np.linalg.solve(A, b)

print(f"La solución es x = {x}")
````

````python
# ~~~~~~~ Ejemplo práctico n.°2 utilizando NumPy ~~~~~~~ #

## Encontramos la parábola que mejor se ajusta a un determinado conjunto de datos.

import numpy as np
import matplotlib.pyplot as plt

x = np.array([0, 1, 2, 3, 4], dtype=float)
y = np.array([1.0, 2.9, 7.2, 12.8, 20.1], dtype=float)

# Ajuste parabólico: y = a x^2 + b x + c
a, b, c = np.polyfit(x, y, 2)
p = np.poly1d([a, b, c])

# Curva suave para dibujar
x_curve = np.linspace(x.min() - 0.5, x.max() + 0.5, 200)
y_curve = p(x_curve)

plt.figure()
plt.scatter(x, y, label="Datos")
plt.plot(x_curve, y_curve, label="Ajuste cuadrático")
plt.title("Ajuste parabólico")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

print(f"Coeficientes: a={a:.6f}, b={b:.6f}, c={c:.6f}")
````

## SymPy

**1) Propósito principal**  
El propósito principal de `SymPy` es realizar **cálculos matemáticos de forma simbólica**.

A diferencia de librerías como `NumPy`, que trabajan con números (valores aproximados), `SymPy` trabaja con las **expresiones matemáticas exactas**, tal como lo harías a mano en un pizarrón.

Esto te permite:

  * **Resolver ecuaciones algebraicamente**: Puedes obtener soluciones exactas como `x = 1/2` o `x = √2` en lugar de `x = 0.5` o `x = 1.414`.
  * **Realizar cálculos de cálculo diferencial e integral**: Permite calcular derivadas e integrales de funciones de manera simbólica, mostrando los pasos y la solución final en su forma de expresión.
  * **Simplificar expresiones**: Puede tomar una expresión matemática compleja y reducirla a su forma más simple.
  * **Trabajar con variables y símbolos**: Mantiene las variables como símbolos, permitiendo manipulaciones algebraicas complejas sin necesidad de asignarles un valor numérico.

En resumen, SymPy es una **Biblioteca de Álgebra Computacional** (CAS, por sus siglas en inglés) que convierte a Python en una poderosa herramienta para el álgebra, el cálculo y otras áreas de las matemáticas que requieren manipulación exacta de fórmulas y símbolos.

**2) Estructuras/objetos que introduce**  
SymPy introduce una familia de objetos diseñados para representar **conceptos matemáticos de forma simbólica y exacta**, no un único contenedor de datos como los arrays.

Los más importantes son:

* Símbolos y Expresiones

  * **Símbolo (`Symbol`):** Es la estructura más elemental. Representa una variable o incógnita matemática (como `x`, `y`, `z`) sin un valor numérico asignado.
    ```python
    x = sp.Symbol('x')
    ```

  * **Expresión simbólica (`Expr`):** Es el objeto principal. Se crea al combinar símbolos con operaciones matemáticas. Representa una fórmula completa (como `x**2 + 2*x + 1`), que puede ser manipulada algebraicamente.
    ```python
    expresion = x**2 + 1
    ```

* Ecuaciones

  * **Ecuación (`Eq`):** Este objeto representa una igualdad matemática. Es fundamental para tareas como resolver ecuaciones, ya que distingue entre una expresión (`x**2 - 1`) y una declaración de igualdad (`x**2 = 1`).
    ```python
    ecuacion = sp.Eq(x**2, 1)
    ```

* Matrices Simbólicas

  * **Matriz (`Matrix`):** A diferencia de los arrays de NumPy que solo contienen números, las matrices de SymPy pueden incluir **símbolos y expresiones**. Esto permite realizar operaciones de álgebra lineal de manera simbólica, como calcular un determinante que resulta en una nueva expresión.
    ```python
    matriz_simbolica = sp.Matrix([[1, x], [x, 1]])
    ```

**3) Funciones/métodos útiles**

Funciones clave de SymPy (con mini-ejemplos y tips):

1. **`symbols`** — crea variables simbólicas (con supuestos)

```python
import sympy as sp
x, y = sp.symbols('x y', real=True)
a = sp.symbols('a', positive=True)   # suposiciones ayudan con raíces, log, etc.
```

2. **`Eq`** — construye ecuaciones simbólicas (≠ usar `==` de Python)

```python
sp.Eq(x + 2, y)      # representa x+2 = y
```

Úsalo con `solve`, `dsolve`, etc.

3. **`simplify`** — simplificación “inteligente” general

```python
sp.simplify((x**2 - 1)/(x - 1))   # -> x + 1
```

Relacionado: `cancel`, `together`, `apart` (fracciones), `trigsimp` (trigonometría).

4. **`factor`**, **`expand`**, **`collect`** — álgebra de polinomios

```python
sp.factor(x**2 - 1)            # -> (x-1)*(x+1)
sp.expand((x + 1)**3)          # -> x**3 + 3x**2 + 3x + 1
sp.collect(3*x**2 + 2*x + x**2, x)  # agrupa por potencias de x
```

5. **`diff`** — derivadas (ordinarias y parciales)

```python
f = sp.sin(x)*x**2
sp.diff(f, x)           # 1ra derivada
sp.diff(f, x, 3)        # 3ra derivada
sp.diff(x*y, x)         # parcial wrt x -> y
```

6. **`integrate`** — integrales indefinidas y definidas

```python
sp.integrate(sp.exp(-x), x)                # indefinida
sp.integrate(sp.exp(-x), (x, 0, sp.oo))    # definida -> 1
```

Tip: `Integral(...)` crea el objeto sin evaluar; `doit()` lo evalúa.

7. **`limit`** — límites (con direcciones)

```python
sp.limit(sp.sin(x)/x, x, 0)            # -> 1
sp.limit(1/x, x, 0, dir='+')           # -> +oo
```

8. **`series`** — series de potencias (Taylor/Maclaurin)

```python
sp.series(sp.sin(x), x, 0, 5)   # x - x**3/3! + O(x**5)
sp.series(sp.exp(x), x, 0, 4).removeO()  # quita el término O(...)
```

9. **`solve`** vs **`solveset`** — resolver ecuaciones/sistemas

```python
sp.solve(sp.Eq(x**2 - 5*x + 6, 0), x)   # -> [2, 3]
sp.solveset(sp.sin(x) - sp.Rational(1,2), x, domain=sp.S.Reals)
```

* `solve` devuelve listas/dict (clásico).
* `solveset` trabaja con **conjuntos** y maneja dominios explícitos (recomendado en problemas más formales).

10. **`nsolve`** — raíces **numéricas** (necesita semilla)

```python
sp.nsolve(sp.cos(x) - x, 0.5)   # -> ~0.739085...
```

11. **`dsolve`** — EDOs (con condiciones iniciales)

```python
t = sp.symbols('t')
y = sp.Function('y')
ode = sp.Eq(sp.diff(y(t), t) - y(t), 0)
sol = sp.dsolve(ode, ics={y(0): 1})     # -> y(t) = exp(t)
```

12. **`Matrix`** — matrices y álgebra lineal simbólica

```python
A = sp.Matrix([[1, 2], [3, 4]])
A.det(), A.inv(), A.eigenvals()  # determinante, inversa, autovalores
```

Útil: `linear_eq_to_matrix` para pasar sistemas a forma `A*x = b`.

13. **`lambdify`** — puente a NumPy/numérico rápido

```python
f = sp.lambdify(x, sp.sin(x)*x, 'numpy')   # f acepta arrays de NumPy
```

También soporta `'math'`, `'mpmath'`, etc.

14. **`subs`**, **`evalf`/`N`** — sustitución y evaluación numérica

```python
expr = sp.sqrt(x + 1)
expr.subs(x, 3)          # -> 2
sp.N(sp.pi, 50)          # pi con 50 dígitos
```

15. **`Poly`**, **`roots`** — objetos polinomiales y raíces exactas

```python
P = sp.Poly(x**3 - 6*x**2 + 11*x - 6, x)  # polinomio
P.degree(), P.all_coeffs()                # grado y coeficientes
sp.roots(P)                               # {1:1, 2:1, 3:1}
```

16. **`pprint` / `latex`** — salida bonita o para informes

```python
sp.pprint((x+1)**3)
sp.latex(sp.integrate(sp.sin(x), x))   # para incrustar en LaTeX
```

17. **Assumptions** (supuestos) y **dominios**

```python
n = sp.symbols('n', integer=True, positive=True)
sp.summation(1/n**2, (n, 1, sp.oo))    # usa supuestos para convergencia
```

Mini-workflow de ejemplo:

```python
import sympy as sp
x = sp.symbols('x', real=True)
expr = (x**2 - 1)/(x - 1)

expr_s = sp.simplify(expr)             # x + 1
d = sp.diff(sp.sin(x)*x**2, x)         # derivar

```

**4) Situación práctica**  
Un problema ideal para SymPy es la optimización de una función para encontrar un máximo o un mínimo. Este tipo de problema es común en física, ingeniería, economía y logística.

**La Situación Práctica: Maximizar un Área**

Imagina que un granjero tiene 100 metros de cerca y quiere construir un corral rectangular que tenga la mayor área posible. El problema es encontrar las dimensiones (largo y ancho) óptimas para lograrlo.

**¿Por Qué SymPy es la Herramienta Ideal?**

* Un enfoque numérico (con NumPy) implicaría probar muchos valores: ¿Qué pasa si el largo es 30m? ¿Y 40m? ¿Y 40.5m? Solo obtendrías una aproximación a la mejor respuesta.

* El enfoque simbólico con SymPy nos permite resolver el problema de forma general y encontrar la solución exacta y probada matemáticamente, tal como lo haríamos con lápiz y papel usando cálculo.

````python
# ~~~~~~~ Ejemplo de SymPy ~~~~~~~ #

# 1) Derivada e integral simbólicas
import sympy as sp

# Definimos el símbolo
x = sp.Symbol('x')

# Creamos la expresión simbólica
funcion = x**3 * sp.sin(x)

# Calculamos la derivada usando sp.diff()
derivada = sp.diff(funcion, x)

print(f"La función original es: f(x) = {funcion}")
print(f"La derivada es: f'(x) = {derivada}")

# Usamos el mismo símbolo 'x'
x = sp.Symbol('x')

# Creamos la nueva expresión
funcion_g = sp.cos(x) + 2*x

# Calculamos la integral indefinida (antiderivada)
integral_indefinida = sp.integrate(funcion_g, x)

# Calculamos la integral definida desde 0 hasta π
#    Para esto, pasamos una tupla: (variable, límite_inferior, límite_superior)
integral_definida = sp.integrate(funcion_g, (x, 0, sp.pi))

print(f"La función a integrar es: g(x) = {funcion_g}")
print(f"La integral indefinida es: ∫g(x)dx = {integral_indefinida}")
print(f"La integral definida de 0 a π es: {integral_definida}")

````

````python
# ~~~~~~~ Ejemplos de SymPy ~~~~~~~ #

# 2) Resolver ecuación y usar lambdify
import sympy as sp

# Definimos el símbolo que usaremos
x = sp.Symbol('x')

# Definimos la ecuación (SymPy asume que la expresión se iguala a 0)
ecuacion = x**2 - x - 6

# Resolvemos la ecuación para 'x'
soluciones = sp.solve(ecuacion, x)

print(f"La expresión simbólica es: {ecuacion}")
print(f"Las soluciones son: {soluciones}")

# Ahora, tomamos la misma expresión x²−x−6 y la convertimos en una función rápida y numérica para evaluarla en varios puntos a la vez.

import numpy as np

# Convertimos la expresión simbólica en una función numérica compatible con NumPy
f_numerica = sp.lambdify(x, ecuacion, "numpy")

# Creamos un arreglo de valores para 'x' desde -5 hasta 5
valores_x = np.linspace(-5, 5, 11)

# Evaluamos la función con todos los valores del arreglo a la vez
valores_y = f_numerica(valores_x)

print(f"\nValores de entrada (x): {valores_x}")
print(f"Resultados de la función (y): {valores_y}")
````

````python
# ~~~~~~~ Ejemplos de SymPy ~~~~~~~ #

import sympy as sp

# --- 1. Definir las Ecuaciones ---
# Definimos las variables simbólicas para el largo y el ancho.
l, w = sp.symbols('l w')

# Ecuación del perímetro: 2*l + 2*w = 100
ecuacion_perimetro = sp.Eq(2*l + 2*w, 100)

# Fórmula del área.
area = l * w

# --- 2. Expresar el Área con una Sola Variable ---
# Despejamos 'w' de la ecuación del perímetro.
# sp.solve devuelve una lista, por eso tomamos el primer elemento [0].
w_en_funcion_de_l = sp.solve(ecuacion_perimetro, w)[0]

# Sustituimos 'w' en la fórmula del área para que dependa solo de 'l'.
area_en_funcion_de_l = area.subs(w, w_en_funcion_de_l)

# --- 3. Encontrar el Máximo con Cálculo ---
# Calculamos la derivada del área con respecto a 'l'.
derivada_area = sp.diff(area_en_funcion_de_l, l)

# Igualamos la derivada a cero y resolvemos para 'l' para encontrar el punto crítico.
l_optimo = sp.solve(derivada_area, l)[0]

# --- 4. Calcular las Dimensiones y el Área Máxima ---
# Usamos la expresión para 'w' que encontramos antes para hallar el ancho óptimo.
w_optimo = w_en_funcion_de_l.subs(l, l_optimo)

# Calculamos el área máxima sustituyendo los valores óptimos.
area_maxima = area_en_funcion_de_l.subs(l, l_optimo)


# --- 5. Mostrar los Resultados ---
print("--- Resolución Simbólica del Problema del Corral ---")
print(f"Área en función de una variable: A(l) = {area_en_funcion_de_l}")
print(f"Derivada del área: A'(l) = {derivada_area}")
print("\n--- Dimensiones Óptimas Encontradas --- ")
print(f"Largo óptimo: {l_optimo} metros")
print(f"Ancho óptimo: {w_optimo} metros")
print(f"Área máxima posible: {area_maxima} metros cuadrados")
````

## SciPy

**1) Propósito principal**  
`SciPy` es una biblioteca de computación científica en Python que extiende a `NumPy` con un amplio conjunto de algoritmos y herramientas (p. ej., optimización, integración, interpolación, álgebra lineal, EDOs, procesamiento de señales y estadística) para resolver problemas numéricos de ingeniería y ciencia.

**2) Estructuras/objetos que introduce**  
A diferencia de otras bibliotecas como `NumPy` o `SymPy`, **SciPy no introduce ninguna estructura de datos fundamental nueva**.

El propósito principal de SciPy no es crear nuevos tipos de objetos, sino **operar sobre las estructuras de datos que proporciona NumPy**. La estructura de datos central sobre la que trabajan casi todos los algoritmos de `SciPy` es el **`ndarray` (array N-dimensional) de NumPy**.

Por lo tanto, cuando usas SciPy, importas la biblioteca para acceder a sus funciones y algoritmos, pero los datos que les pasas y que recibes de vuelta son, en su mayoría, arrays de NumPy.

**3) Funciones/métodos útiles**  

A continuación, listaremos algunos de los "submódulos" que tiene `SciPy` y las funciones de cada uno:

1. `scipy.integrate` — Integración y EDOs

* **Problema típico:** integrar funciones y resolver **ecuaciones diferenciales ordinarias (EDO)** de valor inicial.
* **Funciones clave:** `quad`, `dblquad`, `nquad`; **`solve_ivp`** (métodos RK45, BDF, etc.).
* **Ejemplo mínimo:**

  ```python
  from scipy.integrate import solve_ivp
  f = lambda t, y: -0.5*y
  sol = solve_ivp(f, (0, 10), [2.0])
  sol.y[:, -1]  # y(10)
  ```

2. `scipy.optimize` — Optimización, raíces y ajuste

* **Problema típico:** **minimización** (con/sin restricciones), búsqueda de **raíces**, **ajuste** de curvas y **programación lineal**.
* **Funciones clave:** **`minimize`**, `least_squares`, `curve_fit`, `root`, `linprog`.
* **Ejemplo mínimo:**

  ```python
  import numpy as np
  from scipy.optimize import minimize
  f = lambda x: (x[0]-1)**2 + (x[1]+2)**2
  res = minimize(f, x0=np.zeros(2), method="L-BFGS-B")
  res.x
  ```

3. `scipy.interpolate` — Interpolación 1D/ND y splines

* **Problema típico:** **interpolar** datos 1D/ND; construir **splines** y superficies.
* **Funciones clave:** **`CubicSpline`**, `UnivariateSpline`, `griddata`, `Rbf`.
* **Ejemplo mínimo:**

  ```python
  import numpy as np
  from scipy.interpolate import CubicSpline
  x = np.linspace(0, 1, 5); y = np.sin(2*np.pi*x)
  cs = CubicSpline(x, y)
  cs(0.25)
  ```

4. `scipy.linalg` — Álgebra lineal densa

* **Problema típico:** resolver **Ax=b**, autovalores, SVD y **descomposiciones** (QR, Cholesky, …).
* **Funciones clave:** `solve`, `eigh/eig`, `svd`, `cholesky`, `qr`.
* **Ejemplo mínimo:**

  ```python
  import numpy as np
  from scipy.linalg import solve
  A = np.array([[3., 2.], [1., 2.]])
  b = np.array([5., 5.])
  solve(A, b)
  ```

5. `scipy.fft` — Transformadas de Fourier

* **Problema típico:** **FFT/IFFT** 1D/ND, análisis espectral, frecuencias.
* **Funciones clave:** `fft`, `ifft`, `fftn/ifftn`, `rfft/irfft`, `fftfreq`.
* **Ejemplo mínimo:**

  ```python
  import numpy as np
  from scipy.fft import fft, fftfreq
  x = np.linspace(0, 1, 256, endpoint=False)
  y = np.sin(2*np.pi*10*x)
  ```

**4) Situación práctica**  
El Problema: **Encontrar la Ruta Más Corta**

El problema fundamental es encontrar el camino de menor costo entre dos puntos. Esto es algo que aplicaciones como Google Maps hacen todo el tiempo. La "ruta más corta" no siempre significa la menor distancia, sino que puede ser el menor tiempo, el menor esfuerzo, etc.

1. Ejemplo Simple: Un Mapa de Ciudades

Para entender el concepto, imaginemos un mapa simple con 4 ciudades (A, B, C, D). Conocemos la distancia directa por carretera entre algunas de ellas. Queremos encontrar la ruta más corta desde la ciudad A a todas las demás.

Este es un problema de grafos clásico. Las ciudades son los nodos y las carreteras son las aristas con un "peso" (la distancia).

**5) Fuentes**  
- Submódulos SciPy: https://docs.scipy.org/doc/scipy/reference/main_namespace.html
- Manual general de SciPy: https://docs.scipy.org/doc/scipy/  
- `solve_ivp`: https://docs.scipy.org/doc/scipy/reference/generated/scipy.integrate.solve_ivp.html  
- Tutorial de integración (EDO): https://docs.scipy.org/doc/scipy/tutorial/integrate.html  
- `minimize`: https://docs.scipy.org/doc/scipy/reference/generated/scipy.optimize.minimize.html  
- Tutorial de optimización: https://docs.scipy.org/doc/scipy/tutorial/optimize.html


````python
# ~~~~~~ Ejemplo práctico con SciPy ~~~~~~ #

import numpy as np
from scipy.sparse import csr_matrix
from scipy.sparse.csgraph import dijkstra

# Matriz de adyacencia: distancias entre 4 ciudades (A, B, C, D)
#       A  B  C  D
#    A [0, 1, 5, 0]  (A->B=1, A->C=5)
#    B [1, 0, 0, 2]  (B->D=2)
#    C [5, 0, 0, 3]  (C->D=3)
#    D [0, 2, 3, 0]
distancias = np.array([
    [0, 1, 5, 0],
    [1, 0, 0, 2],
    [5, 0, 0, 3],
    [0, 2, 3, 0]
])

# Convertimos la matriz a un formato disperso (eficiente para grafos)
grafo_ciudades = csr_matrix(distancias)

# Ejecutamos el algoritmo de Dijkstra desde la ciudad A (índice 0)
distancias_minimas = dijkstra(csgraph=grafo_ciudades, directed=False, indices=0)

print("Distancias mínimas desde la ciudad A:")
print(f"A -> A: {distancias_minimas[0]}")
print(f"A -> B: {distancias_minimas[1]}") # Ruta directa A->B
print(f"A -> C: {distancias_minimas[2]}") # Ruta A->B->D->C (1+2+3=6) es peor que A->C (5)
print(f"A -> D: {distancias_minimas[3]}") # Ruta A->B->D (1+2=3)
````

## Matplotlib

**1) Propósito principal**  
El propósito principal de Matplotlib es crear visualizaciones y gráficos estáticos, animados e interactivos en Python.

Es la biblioteca fundamental para la visualización de datos en el ecosistema científico de Python. Permite convertir datos numéricos (generalmente de arrays de NumPy o DataFrames de Pandas) en una amplia variedad de gráficos, como líneas, barras, histogramas, diagramas de dispersión y muchos más.

**2) Estructuras/objetos que introduce**  
Matplotlib introduce una jerarquía de objetos que, trabajando juntos, construyen un gráfico completo. Los dos objetos más importantes que debes entender son la **`Figure`** y los **`Axes`**.

Piensa en ello con esta analogía:
* La **`Figure`** es el lienzo o la hoja de papel en blanco. 🖼️
* Los **`Axes`** son el área de dibujo específica sobre ese lienzo donde trazas tus datos. 📈

**Los Componentes Principales**

Aquí están las estructuras fundamentales que Matplotlib utiliza, ordenadas de la más grande a la más pequeña:

1. `Figure` (La Figura)

  Es el **contenedor de más alto nivel**. Representa la ventana o página completa en la que todo se dibuja. Una `Figure` puede contener uno o más `Axes` (sub-gráficos).

* **¿Qué controla?**: El tamaño general de la ventana, el color de fondo fuera de los gráficos, y títulos y leyendas que aplican a toda la figura.
* **¿Cómo se crea?**: Generalmente con `plt.figure()` o, más comúnmente, junto con los `Axes` usando `fig, ax = plt.subplots()`.

2. `Axes` (Los Ejes / El Gráfico)

  Este es el objeto con el que más interactuarás. **Es el gráfico en sí mismo**: el área rectangular donde los datos se visualizan con líneas, barras, puntos, etc. Su nombre (`Axes`) es en plural porque se refiere al conjunto de ejes X, Y (y Z en 3D), no a un solo eje.

* **¿Qué contiene?**: La representación visual de los datos (las líneas, barras), los ejes X e Y (`Axis`), el título del gráfico, la cuadrícula (`grid`), y las etiquetas de los ejes.
* **Importante**: No lo confundas con `Axis`. Una `Figure` puede tener múltiples `Axes`, como en un panel con varios gráficos.

  *Una sola `Figure` puede contener múltiples `Axes`, cada uno siendo un gráfico independiente.*

3. `Axis` (El Eje)

  Estos son los objetos tipo "línea numérica" que controlan los límites del gráfico y las marcas (ticks). Cada objeto `Axes` tiene un `xaxis` y un `yaxis`.

* **¿Qué controla?**: Los límites de los datos (ej: de 0 a 100 en el eje Y), la posición de las marcas (`ticks`), las etiquetas de esas marcas (`tick labels`) y la etiqueta del eje (`xlabel`, `ylabel`).
* **¿Cómo se usa?**: Generalmente se accede a ellos a través de los métodos del objeto `Axes`, como `ax.set_xlabel()` o `ax.set_xlim()`.

4. `Artist` (El Artista) 🎨

  Este es un concepto clave: **prácticamente todo lo que ves en una `Figure` es un `Artist`**. Esto incluye la propia `Figure`, los `Axes`, las líneas, el texto, las leyendas, etc.

* **¿Qué son?**: Son los objetos que saben cómo dibujarse a sí mismos en el lienzo.
* **Ejemplos**:
    * `Line2D`: El objeto que representa una línea en un gráfico.
    * `Rectangle`: El objeto que representa una barra en un diagrama de barras.
    * `Text`: El objeto para los títulos, etiquetas y anotaciones.

  Normalmente no manipulas los `Artist` directamente, sino que las funciones como `ax.plot()` o `ax.set_title()` los crean y gestionan por ti.

**Resumen de la Jerarquía**

La estructura se puede resumir así:

* **`Figure`** (El lienzo completo)
    * **`Axes`** (Un gráfico o sub-gráfico dentro del lienzo)
        * **`Axis`** (El eje X y el eje Y, que controlan los límites y las marcas)
        * **`Artist`s** (Las líneas, puntos, textos, leyendas, etc., que se dibujan dentro del `Axes`)

**3) Funciones/métodos útiles**  
Matplotlib introduce funciones para crear y personalizar gráficos, que se pueden usar a través de dos interfaces principales: una basada en el módulo `pyplot` (más simple y rápida para gráficos sencillos) y una interfaz orientada a objetos (más potente y flexible para gráficos complejos).

1\. Interfaz `pyplot` (Estilo MATLAB)

Este conjunto de funciones opera sobre una figura y ejes globales. Es ideal para la exploración rápida de datos.

**Creación de Gráficos Básicos**

  * **`plt.plot()`**: Dibuja un gráfico de líneas.

    ```python
    import matplotlib.pyplot as plt
    plt.plot([0, 1, 2, 3], [10, 20, 5, 15]) # plt.plot(x, y)
    ```

  * **`plt.scatter()`**: Crea un diagrama de dispersión de puntos.

    ```python
    plt.scatter([1, 2, 3, 4], [12, 18, 7, 13]) # plt.scatter(x, y)
    ```

  * **`plt.bar()`**: Genera un gráfico de barras verticales.

    ```python
    plt.bar(['A', 'B', 'C'], [20, 35, 30]) # plt.bar(categorias, valores)
    ```

  * **`plt.hist()`**: Dibuja un histograma para ver la distribución de un conjunto de datos.

    ```python
    datos = [1, 2, 2, 3, 3, 3, 4, 4, 5]
    plt.hist(datos, bins=5) # plt.hist(datos, numero_de_contenedores)
    ```

  * **`plt.imshow()`**: Muestra datos como una imagen (por ejemplo, una matriz).

    ```python
    matriz = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
    plt.imshow(matriz, cmap='viridis')
    ```

**Personalización y Visualización**

  * **`plt.title()`, `plt.xlabel()`, `plt.ylabel()`**: Añaden un título y etiquetas a los ejes.

    ```python
    plt.plot([1, 2, 3], [4, 5, 1])
    plt.title("Mi Gráfico")
    plt.xlabel("Eje X")
    plt.ylabel("Eje Y")
    ```

  * **`plt.legend()`**: Muestra la leyenda del gráfico (requiere la etiqueta `label` en las funciones de trazado).

    ```python
    plt.plot([0, 1], [1, 2], label="Línea 1")
    plt.plot([0, 1], [2, 1], label="Línea 2")
    plt.legend()
    ```

  * **`plt.grid()`**: Agrega una cuadrícula al fondo del gráfico.

    ```python
    plt.plot([1, 2, 3], [2, 3, 2])
    plt.grid(True)
    ```

  * **`plt.show()`**: Muestra la figura final. Es la función que renderiza y abre la ventana del gráfico.

    ```python
    plt.plot([1, 2], [1, 2])
    plt.show()
    ```

2\. Interfaz Orientada a Objetos

Este enfoque te da control explícito sobre la figura (`Figure`) y los ejes (`Axes`). Es el método recomendado para gráficos complejos o personalizados.

Primero, se crea una figura y uno o varios ejes:

```python
fig, ax = plt.subplots() # fig es el lienzo, ax es el gráfico
```

Luego, se usan los métodos del objeto `ax`:

**Creación de Gráficos (Métodos de `Axes`)**

  * **`ax.plot()`**, **`ax.scatter()`**, **`ax.bar()`**, etc.: Son los equivalentes a las funciones de `pyplot`.
    ```python
    fig, ax = plt.subplots()
    ax.scatter([1, 2, 3], [3, 1, 4]) # Llama al método sobre el objeto 'ax'
    ```

**Personalización (Métodos `set_*`)**

  * **`ax.set_title()`, `ax.set_xlabel()`, `ax.set_ylabel()`**: Configuran las etiquetas y el título.

    ```python
    fig, ax = plt.subplots()
    ax.plot([1, 2, 3], [4, 5, 1])
    ax.set_title("Mi Gráfico (Objeto)")
    ax.set_xlabel("Eje X")
    ax.set_ylabel("Eje Y")
    ```

  * **`ax.legend()`** y **`ax.grid()`**: Métodos equivalentes para la leyenda y la cuadrícula.

    ```python
    fig, ax = plt.subplots()
    ax.plot([0, 1], [1, 2], label="Línea 1")
    ax.plot([0, 1], [2, 1], label="Línea 2")
    ax.legend()
    ax.grid(True)
    ```

En resumen, Matplotlib introduce un conjunto de funciones de trazado y personalización que pueden ser llamadas directamente desde `pyplot` para rapidez, o como métodos de objetos `Axes` para un mayor control y flexibilidad.

**4) Situación práctica**  
Una situación práctica y muy común donde Matplotlib es esencial es en el **análisis de datos de series temporales**, como puede ser la evolución de la temperatura a lo largo de un año en una ciudad.

El Problema: Analizar las Temperaturas Anuales en Resistencia

**La Situación:** Un climatólogo quiere visualizar y comunicar los patrones de temperatura de Resistencia, Chaco, para entender mejor las tendencias estacionales. Quiere responder dos preguntas:

1.  ¿Cómo varía la temperatura promedio a lo largo de los meses?
2.  ¿Qué tan grande es la fluctuación de temperaturas diarias dentro de un mes de verano, como enero?

**Por qué Matplotlib es la herramienta ideal:** Un gráfico es la forma más clara y efectiva de mostrar tendencias y variabilidad en datos numéricos. Matplotlib permite crear un panel de visualización personalizado para responder a ambas preguntas de manera simultánea y profesional.

**La Solución con Matplotlib**

Resolveremos este problema creando una figura que contenga dos subgráficos (`Axes`): uno para la tendencia anual y otro para el detalle diario de enero.

**Conclusión del Análisis**

El gráfico generado resuelve el problema de manera efectiva:

1.  **El primer subgráfico (líneas azules)** muestra claramente la curva estacional: un pico de calor en los meses de verano (diciembre-febrero) y un valle más fresco en invierno (junio-julio). La tendencia es inmediatamente comprensible.
2.  **El segundo subgráfico (barras naranjas)** responde a la segunda pregunta al visualizar la variabilidad diaria. Se puede ver cómo las temperaturas máximas de enero fluctúan día a día en torno al promedio mensual (línea roja punteada).

Este ejemplo práctico demuestra cómo Matplotlib no solo crea gráficos, sino que funciona como una potente herramienta de análisis y comunicación de resultados, permitiendo combinar diferentes vistas de los datos en un panel coherente e informativo.

````python
# ~~~~~~ Ejemplo práctico con Matplotlib ~~~~~~ #

import matplotlib.pyplot as plt
import numpy as np

# --- 1. Preparar los Datos (Simulados para el ejemplo) ---

# Datos anuales: Temperaturas promedio mensuales para Resistencia, Chaco.
meses = ['Ene', 'Feb', 'Mar', 'Abr', 'May', 'Jun', 'Jul', 'Ago', 'Sep', 'Oct', 'Nov', 'Dic']
temp_promedio_mensual = [28, 27, 26, 22, 19, 17, 16, 18, 20, 23, 25, 27]

# Datos diarios: Temperaturas máximas diarias simuladas para enero.
dias_enero = np.arange(1, 32)
# Simulamos una fluctuación realista alrededor del promedio de 28°C.
np.random.seed(42)
temp_diaria_enero = 28 + np.random.randn(31) * 2.5

# --- 2. Crear la Visualización con Matplotlib ---

# Creamos una figura y un conjunto de subgráficos (2 filas, 1 columna).
# `figsize` ajusta el tamaño total para una mejor visualización.
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 8))

# --- Gráfico 1: Tendencia Anual ---
ax1.plot(meses, temp_promedio_mensual, marker='o', linestyle='-', color='b')
ax1.set_title('Temperatura Promedio Mensual en Resistencia')
ax1.set_ylabel('Temperatura (°C)')
ax1.grid(True, linestyle='--', alpha=0.6)

# --- Gráfico 2: Variabilidad Diaria en Enero ---
ax2.bar(dias_enero, temp_diaria_enero, color='coral')
# Añadimos una línea horizontal para marcar el promedio del mes.
ax2.axhline(y=np.mean(temp_diaria_enero), color='r', linestyle='--', label=f'Promedio Enero: {np.mean(temp_diaria_enero):.1f}°C')
ax2.set_title('Variabilidad de la Temperatura Máxima Diaria en Enero')
ax2.set_xlabel('Día del Mes')
ax2.set_ylabel('Temperatura (°C)')
ax2.legend()
ax2.set_ylim(20, 35) # Ajustamos el eje Y para ver mejor la fluctuación.

# --- 3. Ajustes Finales y Muestra ---

# Añadimos un título general a toda la figura.
fig.suptitle('Análisis Climático de Resistencia, Chaco', fontsize=16)

# Ajusta el espaciado entre los gráficos para que no se superpongan.
fig.tight_layout(rect=[0, 0.03, 1, 0.95])

# Muestra el gráfico final.
plt.show()
````

## 1) Derivadas simbólicas (diff): simples y parciales

**¿Cómo se calculan?** Con `sympy.diff(expr, var, n)` donde `n` es el orden (opcional). Para derivadas parciales, se usa la misma función indicando la variable correspondiente.  
Ejemplos y sintaxis se muestran en los tutoriales oficiales de SymPy sobre **cálculo**. Ver: [Tutorial de Cálculo](https://docs.sympy.org/latest/tutorials/intro-tutorial/calculus.html).

**Ejemplo (derivada simple):**

- $f(x)=x^{3}\sin x \;\Rightarrow\; f'(x)=3x^{2}\sin x + x^{3}\cos x.$

**Ejemplo (derivadas parciales):**

- $g(x,y)=x^{2}y + \sin(xy) \;\Rightarrow\; \frac{\partial g}{\partial x}=2xy + y\cos(xy),\; \frac{\partial g}{\partial y}=x^{2} + x\cos(xy).$



````python
# Derivadas con sympy.diff
f = x**3*sp.sin(x)
df = sp.diff(f, x, 1)           # primera derivada
d3f = sp.diff(f, x, 3)       # tercera derivada

df, d3f
````

````python
# Derivadas parciales
g = x**2*y + sp.sin(x*y)
dg_dx = sp.diff(g, x, 1)
dg_dy = sp.diff(g, y, 1)

# mixta de segundo orden
d2g_dxdy = sp.diff(g, x, y)

dg_dx, dg_dy, d2g_dxdy
````

## 2) Cálculo de límites simbólicos (limit)

Usá `sympy.limit(expr, var, valor, dir='+')` para límite por derecha (o `dir='-'` por izquierda).  
El módulo de **series** explica el algoritmo de límites (basado en la idea de **Gruntz**). Ver:
- [Tutorial de Cálculo (límites)](https://docs.sympy.org/latest/tutorials/intro-tutorial/calculus.html)  
- [Módulo de series / límites](https://docs.sympy.org/latest/modules/series/series.html)

**Ejemplo clásico:** \\( \lim_{x\to 0} \frac{\sin x}{x} = 1 \\)
  
**Unilaterales:** \\( \lim_{x\to 0^+} \frac{|x|}{x} = 1 \\) y \\( \lim_{x\to 0^-} \frac{|x|}{x} = -1 \\).


````python
# Límites
L1 = sp.limit(sp.sin(x)/x, x, 0)

L_right = sp.limit(sp.Abs(x)/x, x, 0, dir='+')
L_left  = sp.limit(sp.Abs(x)/x, x, 0, dir='-')

L1, L_right, L_left
````

- Puede justificar automáticamente pasos que manualmente serían largos o propensos a errores.  
- Implementa algoritmos simbólicos específicos para límites (p. ej., ideas del **algoritmo de Gruntz**). Ver:  
  - [Series / Límites en SymPy (referencia)](https://docs.sympy.org/latest/modules/series/series.html)

> **Nota:** Como toda herramienta simbólica, tiene **limitaciones** en casos no analíticos o muy patológicos (por ejemplo, funciones a trozos complejas). En tales casos, puede requerir reformular el problema o usar supuestos adicionales.


## 4) ¿Cuándo el límite simbólico es más útil que evaluarlo numéricamente?

**Ejemplo de cancelación catastrófica**

$$
\lim_{x\to 0}\frac{\sqrt{1+x}-1}{x}=\frac{1}{2}.
$$

Si evaluás numéricamente para $x$ muy pequeño, la resta $\,\sqrt{1+x}-1\,$ puede perder precisión.  
El enfoque **simbólico** evita estos errores y devuelve el valor exacto $\tfrac{1}{2}$.  
(Para evaluar numéricamente con estabilidad, **racionalizá**:)

$$
\frac{\sqrt{1+x}-1}{x}
=\frac{(\sqrt{1+x}-1)(\sqrt{1+x}+1)}{x(\sqrt{1+x}+1)}
=\frac{1}{\sqrt{1+x}+1}.
$$


````python
expr = (sp.sqrt(1+x) - 1)/x
L = sp.limit(expr, x, 0)  # valor exacto simbólico

# Aproximaciones numéricas ingenuas
vals = [10**(-k) for k in range(1,8)]
approx = [(v, float(expr.subs(x, v))) for v in vals]

L, approx
````

## 2) ¿Cuál es la diferencia entre `matplotlib.pyplot` y otras formas de uso?

La diferencia principal entre `matplotlib.pyplot` y otras formas de usar Matplotlib radica en el **paradigma de programación** y el **nivel de control** que ofrecen.  `pyplot` es una interfaz más simple y orientada a la creación rápida de gráficos, mientras que otras formas de usar Matplotlib ofrecen mayor flexibilidad y control granular.

Aquí hay un desglose más detallado:

**1. `matplotlib.pyplot` (Interfaz de estado):**

*   **Paradigma:** Orientado a estados (stateful).  Esto significa que `pyplot` mantiene una figura y un conjunto de ejes "activos" implícitos. Cuando usas funciones como `plt.plot()`, `plt.title()`, o `plt.xlabel()`, estás modificando esos objetos activos.
*   **Simplicidad:** Es la forma más común de interactuar con Matplotlib, especialmente para usuarios principiantes.  Es fácil de aprender y usar para la creación rápida de gráficos básicos.
*   **Funcionalidad:** Ofrece funciones de alto nivel para la mayoría de las tareas de visualización comunes.
*   **Flexibilidad limitada:**  Aunque es posible personalizar gráficos usando `pyplot`, el control directo sobre los objetos subyacentes (Figura, Axes, etc.) es limitado.  Es más difícil crear gráficos complejos o manipular objetos individuales.

**2. Enfoque orientado a objetos (OOP):**

*   **Paradigma:** Orientado a objetos (object-oriented).  En este enfoque, creas explícitamente objetos `Figure` y `Axes` y luego llamas a métodos sobre estos objetos para crear y modificar el gráfico.
*   **Control:** Ofrece un control mucho más granular sobre cada aspecto del gráfico. Puedes acceder y modificar las propiedades de cada elemento de la figura.
*   **Flexibilidad:** Es ideal para crear gráficos complejos, integrar Matplotlib en aplicaciones más grandes y personalizar el comportamiento de la biblioteca.
*   **Complejidad:** Requiere una comprensión más profunda de la jerarquía de objetos de Matplotlib (Figure, Axes, Axis, etc.).


**3. `matplotlib.figure.Figure` y `matplotlib.axes._axes.Axes` (Acceso Directo):**

*   Esta es una manera aún más avanzada de usar Matplotlib, donde interactúas directamente con las clases `Figure` y `Axes` (y otras clases como `Line2D`, `Text`, etc.) para controlar cada aspecto del gráfico.
*   Es el enfoque más flexible, pero también el más complejo.  Generalmente, no es necesario usar este enfoque directamente a menos que estés creando visualizaciones muy especializadas o contribuyendo al desarrollo de la propia biblioteca Matplotlib.

**Resumen en Tabla:**

| Característica          | `matplotlib.pyplot`       | Enfoque Orientado a Objetos |
|--------------------------|--------------------------|-----------------------------|
| Paradigma              | Orientado a estados       | Orientado a objetos         |
| Facilidad de uso       | Muy fácil                 | Moderado                   |
| Control                  | Limitado                 | Mayor                      |
| Flexibilidad             | Limitada                 | Mayor                      |
| Curva de aprendizaje     | Baja                     | Moderada                   |
| Tipos de gráficos       | Gráficos básicos         | Gráficos complejos         |
| Mantenimiento de estado | Implícito                | Explícito                  |

**Cuándo usar cada enfoque:**

*   **`pyplot`:** Para análisis exploratorio rápido, prototipado rápido de visualizaciones y creación de gráficos básicos.  Es ideal para empezar.
*   **OOP:** Para gráficos más complejos, personalización avanzada, integración en aplicaciones más grandes y cuando necesitas un control preciso sobre los objetos del gráfico.

````python
# ~~~ Ejemplo de enfoque pyplot ~~~ #

## En este ejemplo, `plt.plot()` crea un gráfico en la figura y los ejes activos, que `pyplot` gestiona internamente.  No estamos creando explícitamente objetos `Figure` o `Axes`.

import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

plt.plot(x, y)
plt.title("Gráfico Lineal")
plt.xlabel("Eje X")
plt.ylabel("Eje Y")
plt.show()
````

````python
# ~~~ Ejemplo de enfoque OOP ~~~ #

## En este ejemplo,  `plt.subplots()` crea explícitamente una figura (`fig`) y un conjunto de ejes (`ax`).  Luego, llamamos a los métodos `plot()`, `set_title()`, etc., directamente en el objeto `ax`.

import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

fig, ax = plt.subplots()  # Crea una figura y un conjunto de ejes. Principal diferencia con el "enfoque pyplot".

ax.plot(x, y)            # Dibuja los datos en los ejes
ax.set_title("Gráfico Lineal") # Establece el título de los ejes
ax.set_xlabel("Eje X")      # Establece la etiqueta del eje X
ax.set_ylabel("Eje Y")      # Establece la etiqueta del eje Y

plt.show()
````

## 3) ¿Qué tipos de gráficos se pueden generar con Matplotlib?

Matplotlib es una biblioteca extremadamente versátil que permite crear una vasta gama de gráficos. Su poder reside en que ofrece control sobre casi todos los aspectos de una figura, lo que posibilita la generación de visualizaciones, desde las más simples hasta las más complejas y personalizadas.

Los tipos de gráficos se pueden agrupar en varias categorías principales:

1. **Gráficos para Comparación y Relación**

  Estos gráficos son ideales para comparar valores entre diferentes categorías o para mostrar la relación entre dos o más variables.

*   **Gráfico de Líneas (`plot`):** Es el gráfico más básico, usado para representar la relación entre dos variables. Es especialmente útil para visualizar tendencias a lo largo del tiempo.
*   **Gráfico de Barras (`bar`, `barh`):** Se utiliza para comparar cantidades entre diferentes grupos o categorías. Pueden ser verticales u horizontales. También existen variantes como los gráficos de barras apiladas o agrupadas.
*   **Gráfico de Dispersión (`scatter`):** Muestra la relación o correlación entre dos variables numéricas. Cada punto representa una observación en el conjunto de datos.
*   **Gráfico de Tallo (`stem`):** Similar a un gráfico de líneas, pero dibuja líneas verticales desde una línea base hasta los puntos de datos.
*   **Gráfico de Escalera (`stairs`):** Crea un gráfico de tipo escalonado, útil para representar funciones que cambian de valor en puntos discretos.

2. **Gráficos para Visualizar Distribuciones**

  Estos gráficos ayudan a entender cómo se distribuyen los datos, su frecuencia y sus características estadísticas.

*   **Histograma (`hist`):** Representa la distribución de una única variable numérica agrupando los datos en "bins" o intervalos y mostrando la frecuencia de observaciones en cada uno.
*   **Diagrama de Cajas o Box Plot (`boxplot`):** Muestra la distribución de los datos a través de sus cuartiles. Es muy eficaz para identificar la mediana, el rango intercuartílico y los valores atípicos (outliers).
*   **Gráfico de Violín (`violinplot`):** Combina un diagrama de cajas con una estimación de la densidad de los datos. No solo muestra los cuartiles, sino también la forma completa de la distribución de los datos.
*   **Gráfico de Eventos (`eventplot`):** Muestra datos idénticos o eventos en función de una posición. Es útil para visualizar la distribución de datos categóricos a lo largo de un eje.

3. **Gráficos para Representar Partes de un Todo**

*   **Gráfico Circular o de Pastel (`pie`):** Muestra la proporción de diferentes categorías como porciones de un círculo. Es útil para representar porcentajes de un total.
*   **Gráfico de Áreas Apiladas (`stackplot`):** Similar a un gráfico de líneas, pero con el área debajo de las líneas rellena de color. Es útil para mostrar cómo cambia la composición de un todo a lo largo del tiempo.

4. **Gráficos Bidimensionales y de Campos**

  Estos gráficos son útiles para visualizar datos en una matriz o en un espacio de dos dimensiones.

*   **Mapa de Calor (`imshow`, `pcolormesh`):** Representa valores de una matriz mediante colores. Es muy utilizado para visualizar matrices de correlación o datos geográficos.
*   **Gráfico de Contorno (`contour`, `contourf`):** Se utiliza para visualizar datos en 3D en un formato 2D, dibujando líneas de contorno que conectan puntos de igual valor (isopletas).
*   **Gráfico de Flechas o de Carcaj (`quiver`):** Muestra la dirección y magnitud de un vector en diferentes puntos, útil en campos como la meteorología o la física.

5. **Gráficos 3D**

  Matplotlib, a través de su toolkit `mplot3d`, también ofrece capacidades para la visualización en tres dimensiones.

*   **Gráficos de Líneas y Dispersión 3D (`plot`, `scatter`):** Extensiones de sus contrapartes 2D para visualizar datos en un espacio tridimensional.
*   **Superficies 3D (`plot_surface`):** Para representar funciones de dos variables (z = f(x, y)).
*   **Gráficos de Barras 3D (`bar3d`):** Versión tridimensional de los gráficos de barras.
*   **Gráficos de Malla (`plot_wireframe`):** Dibuja una superficie utilizando únicamente una malla de líneas.

6. **Gráficos Polares**

  También es posible crear gráficos utilizando un sistema de coordenadas polares en lugar del cartesiano.

*   **Gráfico de Barras Polar (`bar` en eje polar)**
*   **Gráfico de Líneas Polar (`plot` en eje polar)**

## 5) Dos funciones clave y qué hacen

1.**`matplotlib.pyplot.subplots()` (usada como `plt.subplots()`)**

Esta función es el punto de partida preferido para crear cualquier gráfico en Matplotlib.

**¿Qué hace?**
De una sola vez, `plt.subplots()` **crea una figura (el lienzo completo) y un conjunto de subgráficos (los ejes o áreas de dibujo) dentro de esa figura.** Devuelve ambos objetos para que puedas trabajar con ellos directamente.

**Explicación detallada:**
*   **Figura (`fig`):** Es el contenedor de más alto nivel. Piensa en ella como el lienzo en blanco o la ventana completa que contendrá todos tus gráficos, títulos y leyendas.
*   **Ejes (`ax`):** Es el área específica donde se trazan los datos. Tiene sus propias coordenadas X e Y, su propio título, etiquetas, etc. Una figura puede contener uno o múltiples objetos de ejes.

La principal ventaja de `plt.subplots()` es que empaqueta este proceso de creación en una sola llamada y te entrega las variables (`fig` y `ax`) que necesitas para seguir el enfoque orientado a objetos, que es más explícito y potente.

**Parámetros comunes:**
*   `nrows`, `ncols`: Permiten crear una cuadrícula de subgráficos. Por ejemplo, `plt.subplots(nrows=2, ncols=3)` crearía una figura con 6 áreas de gráfico, organizadas en 2 filas y 3 columnas.
*   `figsize`: Una tupla que define el ancho y alto de la figura en pulgadas, por ejemplo, `figsize=(10, 6)`.

**¿Por qué es clave?**
Es la puerta de entrada al método de trabajo más robusto y flexible de Matplotlib. Al usar `fig, ax = plt.subplots()`, inmediatamente obtienes control directo y explícito sobre el lienzo (`fig`) y el área de dibujo (`ax`).

**2. `axes.plot()` (usada como `ax.plot()`)**

Una vez que tienes tu área de dibujo (`ax`), esta es la función principal para trazar datos en ella.

**¿Qué hace?**
La función `ax.plot()` **dibuja un gráfico de líneas conectando una secuencia de puntos de datos (x, y) en el objeto de ejes (`ax`) específico sobre el que se llama.**

**Explicación detallada:**
*   A diferencia de `plt.plot()` (la versión de `pyplot`), que dibuja en el "gráfico activo" que Matplotlib esté gestionando internamente, `ax.plot()` es un método que se aplica a un objeto `ax` que tú controlas. Esto elimina la ambigüedad, especialmente cuando trabajas con múltiples subgráficos.
*   Acepta una gran variedad de datos de entrada, como listas de Python o, más comúnmente, arrays de NumPy.

**Parámetros comunes:**
*   `x`, `y`: Las coordenadas de los puntos a trazar.
*   `color`: Para especificar el color de la línea (ej. `'red'`, `'blue'`).
*   `linestyle`: Para cambiar el estilo de la línea (ej. `'--'` para una línea discontinua, `':'` para una línea de puntos).
*   `marker`: Para poner un marcador en cada punto de datos (ej. `'o'` para círculos, `'x'` para equis).
*   `label`: Un texto que se usará para identificar la línea en una leyenda.

**¿Por qué es clave?**
Es el método fundamental para visualizar las relaciones entre datos numéricos. Mientras que `plt.subplots()` prepara el escenario, `ax.plot()` es el actor principal que pone los datos en ese escenario.


En resumen:
1.  **`plt.subplots()`:** Prepara tu espacio de trabajo (el lienzo y el área de dibujo).
2.  **`ax.plot()`:** Dibuja tus datos en el área de dibujo que preparaste.

````python
# ~~~ Ejemplo con subplots y plot ~~~ #

## Este código muestra cómo estas dos funciones trabajan juntas en perfecta armonía:

import matplotlib.pyplot as plt
import numpy as np

# Datos de ejemplo
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# 1. Usar plt.subplots() para crear la figura y los ejes
#    Esto nos da el control explícito sobre 'fig' y 'ax'
fig, ax = plt.subplots(figsize=(8, 5))

# 2. Usar ax.plot() para dibujar los datos EN los ejes que creamos
#    Dibujamos dos líneas diferentes en el mismo 'ax'
ax.plot(x, y1, color='blue', linestyle='-', label='Seno')
ax.plot(x, y2, color='red', linestyle='--', label='Coseno')

# Ahora usamos otros métodos del objeto 'ax' para personalizar el gráfico
ax.set_title("Función Seno y Coseno")
ax.set_xlabel("Eje X (radianes)")
ax.set_ylabel("Amplitud")
ax.legend()  # Muestra la leyenda usando las 'labels' que definimos
ax.grid(True) # Añade una cuadrícula

# Finalmente, le decimos a pyplot que muestre la figura que hemos construido
plt.show()
````

## Gráfico de una función con el método de **interfaz de estado (stateful interface)**.

El código a continuación utilizará el método "pyplot", también conocido como la interfaz de estado (stateful interface).

¿Cómo lo sabemos?: La clave para identificarlo es que todas las funciones para crear y modificar el gráfico se llaman directamente desde `plt`:

    plt.plot(...)

    plt.title(...)

    plt.xlabel(...)

    plt.ylabel(...)

    plt.grid(...)

    plt.legend()

    plt.show()

En este paradigma, no se crean explícitamente los objetos de la figura (fig) y de los ejes (ax). En su lugar, pyplot mantiene internamente una referencia a una figura y unos ejes "actuales" o "activos". Cada vez que llamas a una función como plt.plot(), esta actúa sobre esos ejes activos.

Piensa en ello como darle órdenes a un asistente: "Dibuja una línea", "Pon un título", "Añade una etiqueta". El asistente sabe en qué gráfico está trabajando en ese momento.

````python
# --- PASO 0: IMPORTACIÓN DE LIBRERÍAS ---
# Importamos la librería NumPy y le asignamos el alias 'np'.
# La usaremos para generar nuestros datos numéricos (los puntos de la función).
import numpy as np

# Importamos el módulo pyplot de la librería Matplotlib y le asignamos el alias 'plt'.
# En este método, usaremos 'plt' para CADA comando relacionado con la creación y personalización del gráfico.
import matplotlib.pyplot as plt


# --- PASO 1: PREPARACIÓN DE LOS DATOS ---
# Creamos un array de NumPy llamado 'x' que contiene 200 puntos.
# Estos puntos están espaciados uniformemente en el intervalo que va desde -10 hasta 10.
# Este array 'x' representa el dominio de nuestra función (eje horizontal).
x = np.linspace(-10, 10, 200)

# Creamos un segundo array llamado 'y'.
# Cada elemento de 'y' es el resultado de elevar al cubo el elemento correspondiente en 'x'.
# Este array 'y' representa el rango de nuestra función (eje vertical).
y = x**3


# --- PASO 2: CREACIÓN DEL GRÁFICO (El método Pyplot) ---
# Esta es la función principal en este enfoque.
# Al llamar a plt.plot(), Matplotlib crea automáticamente una figura y unos ejes "activos" en segundo plano.
# Todas las funciones siguientes modificarán este gráfico "activo".
plt.plot(
    x, y,               # Los datos a graficar: coordenadas 'x' y coordenadas 'y'.
    color="red",        # Argumento para definir el color de la línea.
    linestyle="--",     # Argumento para que la línea sea discontinua (con guiones).
    linewidth=2,        # Argumento para establecer el grosor de la línea en 2 puntos.
    label="f(x) = x³"   # Asigna una etiqueta a esta línea, que se usará para la leyenda.
)


# --- PASO 3: PERSONALIZACIÓN DEL GRÁFICO ---
# A partir de aquí, cada función llamada desde 'plt' modifica el gráfico que se acaba de crear.
# Agregamos un título al gráfico "activo" usando la función plt.title().
plt.title("Gráfico de la función cúbica f(x) = x³") # Corregido: es cúbica, no cuadrática

# Establecemos la etiqueta para el eje horizontal usando la función plt.xlabel().
plt.xlabel("Eje X")

# Establecemos la etiqueta para el eje vertical usando la función plt.ylabel().
plt.ylabel("Eje Y")

# Activamos una cuadrícula de fondo para mayor claridad con la función plt.grid().
# 'True' la activa, 'linestyle=":"' la hace punteada y 'alpha=0.5' le da un 50% de transparencia.
plt.grid(True, linestyle=":", alpha=0.5)

# La función plt.legend() busca automáticamente cualquier línea con una 'label' y muestra la leyenda.
plt.legend()


# --- PASO 4: MOSTRAR EL GRÁFICO FINAL ---
# Finalmente, usamos la función plt.show().
# Esta función busca el gráfico que hemos estado construyendo en memoria y lo renderiza en una ventana en la pantalla.
plt.show()
````

## Gráfico de una función utilizando el método **orientado a objetos (OOP)**

Antes de ver el código, es importante entender el **método orientado a objetos**. Imagina que en lugar de darle órdenes generales a un "asistente" (`plt`), ahora pides tus propias herramientas de trabajo y las controlas tú mismo.

Pensemos en ello como ser un pintor:

1.  **Pides tu Lienzo y tu Área de Dibujo:** En lugar de empezar a pintar al aire, primero pides un lienzo en blanco y defines el área donde vas a dibujar. En Matplotlib, esto se hace con una línea clave:
    
    ```python
    fig, ax = plt.subplots()
    ```
    *   **`fig` (la Figura):** Es tu lienzo completo. El marco que lo contiene todo.
    *   **`ax` (los Ejes):** Es tu área de dibujo específica dentro de ese lienzo. Aquí es donde realmente irán las líneas, las barras y los puntos.

2.  **Trabajas Directamente sobre tus Herramientas:** Una vez que tienes tu lienzo (`fig`) y tu área de dibujo (`ax`), todas las órdenes se las das directamente a ellos.
    *   En lugar de decir `plt.plot()` (orden general: "dibuja esto"), ahora dices `ax.plot()` (orden específica: "**en esta área de dibujo**, traza esta línea").
    *   En lugar de `plt.title()`, ahora usas `ax.set_title()` ("**a esta área de dibujo**, ponle este título").

La gran ventaja es el **control total y la claridad**. Sabes exactamente qué elemento del gráfico estás modificando en cada momento. Esto es fundamental cuando quieres crear gráficos más complejos o tener varios subgráficos en una misma figura.

En resumen: **creamos los elementos del gráfico (`fig`, `ax`) primero y luego los modificamos directamente.**

Veamos cómo dibujar el gráfico anterior (y = x³) con el método OOP:

````python
# --- PASO 0: IMPORTACIÓN DE LIBRERÍAS ---
# Importamos la librería NumPy y le asignamos el alias 'np'.
# NumPy es fundamental para trabajar con arrays (vectores y matrices) y realizar cálculos matemáticos de forma eficiente.
import numpy as np

# Importamos el módulo pyplot de la librería Matplotlib y le asignamos el alias 'plt'.
# Pyplot nos proporciona las funciones para crear la figura y mostrar el gráfico final.
import matplotlib.pyplot as plt


# --- PASO 1: PREPARACIÓN DE LOS DATOS ---
# Creamos un array de NumPy llamado 'x' que contiene 200 puntos.
# Estos puntos están espaciados uniformemente en el intervalo que va desde -10 hasta 10.
# Este array 'x' representa el dominio de nuestra función, es decir, los valores del eje horizontal.
x = np.linspace(-10, 10, 200)

# Creamos un segundo array llamado 'y'.
# Cada elemento de 'y' es el resultado de elevar al cubo el elemento correspondiente en 'x'.
# Este array 'y' representa el rango de nuestra función, es decir, los valores del eje vertical.
y = x**3


# --- PASO 2: CREACIÓN DE LA FIGURA Y LOS EJES (El método OOP) ---
# Esta es la línea central del enfoque orientado a objetos.
# La función plt.subplots() crea y devuelve dos objetos a la vez:
# 1. 'fig': Un objeto Figure, que es el contenedor principal, como si fuera el lienzo completo o la ventana.
# 2. 'ax': Un objeto Axes, que es el área de dibujo real donde se trazarán los datos. Contiene los ejes, las etiquetas, etc.
# Guardamos estos dos objetos en sus respectivas variables para poder controlarlos directamente.
fig, ax = plt.subplots()


# --- PASO 3: DIBUJAR LOS DATOS SOBRE LOS EJES ---
# Llamamos al método .plot() directamente sobre nuestro objeto 'ax'.
# Esto le ordena a Matplotlib que dibuje los datos (x, y) específicamente en esta área de dibujo.
ax.plot(
    x, y,               # Los datos a graficar: coordenadas 'x' y coordenadas 'y'.
    color="red",        # Argumento para definir el color de la línea.
    linestyle="--",     # Argumento para que la línea sea discontinua (con guiones).
    linewidth=2,        # Argumento para establecer el grosor de la línea en 2 puntos.
    label="f(x) = x³"   # Asigna una etiqueta a esta línea, que se usará para crear la leyenda.
)


# --- PASO 4: PERSONALIZACIÓN DEL GRÁFICO ---
# A partir de aquí, modificamos las propiedades del objeto 'ax' para mejorar la visualización.
# Usamos el método .set_title() del objeto 'ax' para establecer el título del gráfico.
ax.set_title("Gráfico de la función cúbica f(x) = x³")

# Usamos el método .set_xlabel() para ponerle una etiqueta al eje horizontal.
ax.set_xlabel("Eje X")

# Usamos el método .set_ylabel() para ponerle una etiqueta al eje vertical.
ax.set_ylabel("Eje Y")

# Llamamos al método .grid() sobre 'ax' para añadir una cuadrícula de fondo.
# 'True' la activa, 'linestyle=":"' la hace punteada y 'alpha=0.5' le da un 50% de transparencia.
ax.grid(True, linestyle=":", alpha=0.5)

# Llamamos al método .legend() sobre 'ax'.
# Esto busca automáticamente cualquier línea que haya sido dibujada con una 'label' y muestra la leyenda.
ax.legend()


# --- PASO 5: MOSTRAR EL GRÁFICO FINAL ---
# Finalmente, usamos la función plt.show() del módulo pyplot.
# Esta función busca la figura ('fig') que hemos estado construyendo en memoria y la renderiza en una ventana en la pantalla.
plt.show()
````

### Comparativa entre los métodos de **interfaz de estado** y **orientación a objetos**

| Característica | Método de Interfaz de Estado | Método OOP |
| :--- | :--- | :--- |
| **Creación** | Implícita. `plt.plot()` crea los ejes si no existen. | Explícita. `fig, ax = plt.subplots()`. |
| **Modificación** | Se usan funciones de `plt` (ej. `plt.title`). | Se usan métodos del objeto `ax` (ej. `ax.set_title`). |
| **Ideal para...** | Gráficos rápidos, sencillos y análisis exploratorio. | Gráficos complejos, personalizables, con múltiples subplots o para integrar en aplicaciones. |


## Dispersión (scatter)

````python
# --- PASO 0: IMPORTACIÓN DE LIBRERÍAS ---
# Importamos la librería NumPy y le asignamos el alias 'np'.
# La usaremos para la generación de números aleatorios.
import numpy as np

# Importamos el módulo pyplot de la librería Matplotlib y le asignamos el alias 'plt'.
# Lo usaremos para crear y mostrar el gráfico.
import matplotlib.pyplot as plt


# --- PASO 1: PREPARACIÓN DE LOS DATOS ---
# Creamos un "generador de números aleatorios" moderno de NumPy.
# La semilla (seed) '123' se utiliza para que la secuencia de números "aleatorios" sea siempre la misma
# cada vez que se ejecuta el código. Esto hace que el resultado sea reproducible.
rng = np.random.default_rng(123)

# Generamos 200 números aleatorios para el eje X.
# Estos números siguen una distribución normal (o gaussiana) con:
# - loc=0: La media de los números será 0.
# - scale=1: La desviación estándar será 1 (define la dispersión de los puntos).
# - size=200: Se generarán 200 puntos.
x = rng.normal(loc=0, scale=1, size=200)

# Generamos los datos para el eje Y.
# Los valores de 'y' se basan en una relación lineal con 'x' (y = 2*x),
# pero le añadimos un "ruido" aleatorio para que los puntos no formen una línea perfecta.
# Este ruido también sigue una distribución normal, pero con una desviación estándar de 0.6.
y = 2*x + rng.normal(scale=0.6, size=200)


# --- PASO 2: CREACIÓN Y PERSONALIZACIÓN DEL GRÁFICO (Método Pyplot) ---
# Creamos explícitamente una nueva figura o lienzo para nuestro gráfico.
# Aunque plt.scatter() podría hacerlo por nosotros, esta es una buena práctica.
plt.figure()

# Usamos la función plt.scatter() para crear un gráfico de dispersión.
# A diferencia de plt.plot(), esta función dibuja puntos individuales en las coordenadas (x, y) en lugar de unirlos con una línea.
# - x, y: Los arrays con las coordenadas de cada uno de los 200 puntos.
# - alpha=0.8: Establece una transparencia del 80% para los puntos. Esto es útil para visualizar
#              dónde se superponen y hay mayor densidad de puntos.
plt.scatter(x, y, alpha=0.8)

# Añadimos un título al gráfico usando la función plt.title().
plt.title('Dispersión')

# Establecemos la etiqueta para el eje horizontal.
plt.xlabel('x')

# Establecemos la etiqueta para el eje vertical.
plt.ylabel('y')

# Activamos una cuadrícula de fondo para facilitar la lectura de las coordenadas.
plt.grid(True)


# --- PASO 3: MOSTRAR EL GRÁFICO FINAL ---
# La función plt.show() renderiza la figura con todos sus elementos y la muestra en una ventana.
plt.show()
````

## Histograma

````python
# --- PASO 0: IMPORTACIÓN DE LIBRERÍAS ---
# Importamos la librería NumPy con el alias 'np' para la generación de datos numéricos.
import numpy as np

# Importamos el módulo pyplot de Matplotlib con el alias 'plt' para crear y mostrar la visualización.
import matplotlib.pyplot as plt


# --- PASO 1: PREPARACIÓN DE LOS DATOS ---
# Creamos un generador de números aleatorios con una semilla fija (0).
# La semilla asegura que la secuencia de números "aleatorios" sea siempre la misma,
# haciendo que el gráfico sea reproducible cada vez que se ejecute el código.
# A continuación, generamos 1000 números aleatorios.
# Estos números se extraen de una distribución normal estándar (gaussiana),
# lo que significa que la mayoría de los valores se agruparán en torno a una media de 0.
data = np.random.default_rng(0).normal(size=1000)


# --- PASO 2: CREACIÓN Y PERSONALIZACIÓN DEL GRÁFICO (Método Pyplot) ---
# Creamos una nueva figura o lienzo en blanco para nuestro histograma.
plt.figure()

# Usamos la función plt.hist() para generar el histograma.
# Esta función es fundamental para visualizar la distribución de un conjunto de datos.
# - El primer argumento ('data') es el array de datos que queremos analizar.
# - El argumento 'bins=30' le indica a Matplotlib que divida el rango completo de valores
#   de los datos en 30 intervalos o "contenedores" del mismo tamaño.
#   Luego, cuenta cuántos de los 1000 puntos de datos caen dentro de cada intervalo.
#   La altura de cada barra en el histograma representa esa cuenta (la frecuencia).
plt.hist(data, bins=30)

# Añadimos un título al gráfico que describe lo que estamos viendo:
# la distribución de datos de una Normal con media 0 y desviación estándar 1.
plt.title('Histograma (Normal ~ N(0,1))')

# Establecemos la etiqueta para el eje horizontal. En un histograma, este eje representa
# los rangos de valores de los datos.
plt.xlabel('Valor')

# Establecemos la etiqueta para el eje vertical. Este eje siempre representa la frecuencia,
# es decir, el número de observaciones que caen en cada intervalo (bin).
plt.ylabel('Frecuencia')


# --- PASO 3: MOSTRAR EL GRÁFICO FINAL ---
# La función plt.show() renderiza la figura con el histograma y todos sus componentes
# y la muestra en una nueva ventana.
plt.show()
````

## Imagen (imshow) — matriz como mapa de colores

````python
# --- PASO 0: IMPORTACIÓN DE LIBRERÍAS ---
# Importamos NumPy con el alias 'np' para realizar las operaciones matemáticas y crear la matriz.
import numpy as np
# Importamos pyplot de Matplotlib con el alias 'plt' para la visualización.
import matplotlib.pyplot as plt


# --- PASO 1: PREPARACIÓN DE LOS DATOS (Creación de la Matriz 2D) ---
# Esta línea crea una matriz 2D (que llamamos 'X') de una manera muy ingeniosa. Vamos a desglosarla:
# 1. np.linspace(0, 3*np.pi, 100): Crea un vector (array 1D) de 100 puntos espaciados uniformemente entre 0 y 3π.
# 2. np.sin(...) y np.cos(...): Se calculan el seno y el coseno para cada uno de esos 100 puntos,
#    lo que nos da dos vectores 1D: 'vector_seno' y 'vector_coseno'.
# 3. np.outer(vector_seno, vector_coseno): Esta es la operación clave. Calcula el "producto exterior".
#    Crea una matriz 2D donde cada elemento (i, j) es el producto del elemento i del 'vector_seno'
#    y el elemento j del 'vector_coseno'.
# El resultado es una matriz 'X' de 100x100, cuyos valores crean un patrón de interferencia ondulatoria.
X = np.outer(np.sin(np.linspace(0, 3*np.pi, 100)), np.cos(np.linspace(0, 3*np.pi, 100)))


# --- PASO 2: CREACIÓN Y PERSONALIZACIÓN DEL GRÁFICO (Método Pyplot) ---
# Creamos una nueva figura o lienzo en blanco para nuestra visualización.
plt.figure()

# Usamos la función plt.imshow() (image show) para visualizar la matriz 2D.
# Esta función toma una matriz y asigna un color a cada celda basado en su valor numérico.
# - El primer argumento ('X') es la matriz 2D que queremos mostrar.
# - origin='lower': Este argumento es importante. Por defecto, imshow coloca el origen de coordenadas (0,0)
#   en la esquina superior izquierda. 'lower' lo cambia a la esquina inferior izquierda, que es el
#   estándar para los gráficos matemáticos.
# - aspect='auto': Permite que los "píxeles" de la imagen no sean cuadrados, ajustándose al tamaño de la figura.
plt.imshow(X, origin='lower', aspect='auto')

# Añadimos un título principal al gráfico.
plt.title('imshow de una matriz 2D con un patrón simple')

# Agregamos una barra de color al lado del gráfico.
# Esta barra es crucial porque actúa como una leyenda, mostrando la correspondencia entre
# los valores numéricos de la matriz 'X' y los colores que se ven en la imagen.
# - El argumento 'label' le pone un título a la propia barra de color.
plt.colorbar(label='Intensidad')


# --- PASO 3: MOSTRAR EL GRÁFICO FINAL ---
# La función plt.show() renderiza la figura con la imagen de la matriz, el título y la barra de color,
# y la muestra en una nueva ventana.
plt.show()
````

````python
# --- PASO 0: IMPORTACIÓN DE LIBRERÍAS ---
# Importamos NumPy con el alias 'np' para crear nuestros datos.
import numpy as np
# Importamos pyplot de Matplotlib con el alias 'plt' para crear y guardar el gráfico.
import matplotlib.pyplot as plt


# --- PASO 1: PREPARACIÓN DE LOS DATOS ---
# Creamos un array 'x' con 50 puntos espaciados uniformemente entre 0 y 1.
x = np.linspace(0, 1, 50)
# Creamos un array 'y' calculando el cuadrado de cada elemento de 'x'.
y = x**2


# --- PASO 2: CREACIÓN DEL GRÁFICO (en memoria) ---
# Creamos una nueva figura en blanco. En este punto, la figura solo existe en la memoria del programa.
plt.figure()

# Dibujamos los datos (x, y) en la figura activa en memoria.
plt.plot(x, y)

# Añadimos un título al gráfico que está en memoria.
plt.title('Guardado en disco')


# --- PASO 3: GUARDAR LA FIGURA EN UN ARCHIVO ---
# Esta es la función clave de este script. En lugar de mostrar el gráfico, lo guardamos en el disco.
# La función plt.savefig() toma la figura "activa" y la exporta a un archivo.
plt.savefig(
    'mi_figura.png',    # El primer argumento es el nombre del archivo, incluyendo la extensión.
                        # Matplotlib usa la extensión (.png) para determinar el formato de guardado.
                        # Otros formatos comunes son .jpg, .svg, .pdf.

    dpi=150,            # 'Dots Per Inch' (Puntos por Pulgada). Este argumento controla la resolución de la imagen.
                        # Un valor más alto (ej. 300) produce una imagen de mayor calidad y tamaño de archivo.
                        # 150 es una buena resolución para presentaciones o uso en la web.

    bbox_inches='tight' # 'Bounding Box Inches'. Este útil argumento recorta automáticamente cualquier
                        # espacio en blanco innecesario alrededor del gráfico, asegurando que la imagen
                        # guardada se ajuste perfectamente al contenido.
)


# --- PASO 4: CERRAR LA FIGURA DE LA MEMORIA ---
# La función plt.close() cierra la figura que acabamos de guardar.
# Esto es una muy buena práctica, especialmente en scripts largos o que generan muchos gráficos,
# ya que libera la memoria que la figura estaba utilizando.
# Si no se cerrara, la figura permanecería en memoria hasta que el script terminara.
plt.close()


# --- PASO 5: CONFIRMACIÓN AL USUARIO ---
# Imprimimos un mensaje en la consola para confirmar que el archivo ha sido creado exitosamente.
print('Archivo generado: mi_figura.png')
````

# Pregunta de reflexión

> *¿Por qué creés que la visualización es importante al trabajar con métodos numéricos? ¿Qué aporta un gráfico que no aporta una tabla de valores?*

La visualización es un componente esencial al trabajar con métodos numéricos porque traduce los resultados en una forma que podemos comprender intuitivamente. Mientras que una tabla de valores nos ofrece datos precisos pero aislados, un gráfico revela el comportamiento global del conjunto.

A través de un gráfico, obtenemos una comprensión clara de la forma de una función: podemos identificar al instante si las tendencias son crecientes o decrecientes, localizar picos y valles, o detectar cambios bruscos.

Además, la visualización es una poderosa herramienta para la depuración. Los errores de cálculo o los valores atípicos se manifiestan como anomalías visuales que son fáciles de detectar, a diferencia de en una tabla donde podrían pasar desapercibidos. Finalmente, un gráfico simplifica enormemente la comunicación de resultados, ya que una imagen puede transmitir de un solo vistazo la información que en una tabla requeriría un análisis mucho más detallado.
